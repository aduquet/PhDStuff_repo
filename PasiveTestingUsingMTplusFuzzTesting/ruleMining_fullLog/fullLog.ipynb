```python
import pandas as pd
import numpy as np
import glob as gl
import pickle
from efficient_apriori import apriori
import warnings
import os
import pathlib
warnings.filterwarnings('ignore')
```

```python
def encoder(df):
    data = df.copy()
    listKey = list(data.keys())


    for key in listKey:
        currentlistKey = list(data.keys())
        if data[key].dtype != object:
            try:
                indexAux = currentlistKey.index(key)
                data.insert(indexAux + 1,key + '_str', 0)

                for index, row in data.iterrows():
                    data.at[index, key + '_str'] = key + '==' + str(int(row[key])) 
                    
            except:
                continue
    return data
```

```python
def encode_str(data_enc):
    listKey = list(data_enc.keys())
    data_str = data_enc.copy()

    for key in listKey:
        if data_enc[key].dtype != object:
            data_str.drop(key, inplace=True, axis=1)

        try:
            if key.find('Component') != -1:
                data_str.drop(key, inplace=True, axis=1)
        
        except:
            continue
    
    return data_str
```

```python
def prepro(data):
    names = data.columns.values
    
    data.drop(columns=['Unnamed: 0'], inplace=True)

    for index, row in data.iterrows():
        auxR1 = row['MR1_checker_str']
        auxR2 = row['MR2_checker_str']
        auxR3 = row['MR3_checker_str']
        auxR4 = row['MR4_checker_str']

        data.at[index, 'MR1_checker_str'] = 'MR1 == ' + auxR1
        data.at[index, 'MR2_checker_str'] = 'MR2 == ' + auxR2
        data.at[index, 'MR3_checker_str'] = 'MR3 == ' + auxR3
        data.at[index, 'MR4_checker_str'] = 'MR4 == ' + auxR4

    return data
```

```python
data = pd.read_csv('/Users/alduck/Documents/PhDStuff_repo/PasiveTestingUsingMTplusFuzzTesting/ruleMining_fullLog/finalLog.csv', index_col= 0)
```

```python
data_enc = encoder(data)
```

```python
finalData = encode_str(data_enc)
```

```python
finalData = prepro(finalData)
```
